```python
import pandas as pd
import glob
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
saved_path = "D:\\code\\proj_ml\\ML_Project\\data\\saved"
saved_files = glob.glob(saved_path + "/*.csv")
```

```python
gold_file = [f for f in saved_files if 'gold' in f][0]
merged_df = pd.read_csv(gold_file)
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

for file in saved_files:
    if 'gold' not in file:
        df = pd.read_csv(file)
        df['Date'] = pd.to_datetime(df['Date'])
        merged_df = pd.merge(merged_df, df, on='Date', how='left')
```
